# Sector Analysis

first we need to extract main sector using the column category_list . the category_list column contains values such
as "biotechnology|health care" - in this "biotech manin category " is the main category of the company which we need
to use extract the main category in new column

```python
#extracting the main category 

new_master.loc[:,"maincategory"] = new_master["category_list"].apply(lambda x :x.split("|")[0])
new_master.head()
```

```python
#now we dosen't need this column in our dataframe so droped..
new_master.drop(["category_list"], axis =1 , inplace= True)
```

now we will read mapping file and merge with maincategory with the corrosponding column

```python
mapping = pd.read_csv("mapping.csv")
mapping.head()
```

```python
mapping.isnull().sum()
```

```python
#remove the missing value 
mapping = mapping[~pd.isnull(mapping["category_list"])]
mapping.head()
```

```python
#converting column to lower case

mapping["category_list"] = mapping["category_list"].str.lower()
new_master["maincategory"] = new_master["maincategory"].str.lower()
mapping.head(20)
```

 if we see carefully , we notice something flasy - there are some sector names alter0tive , this is a not random 
quality issue but rather pattern in some string the NA has been replaced with 0 , we need to fix this 

```python
mapping["category_list"] = mapping["category_list"].apply(lambda x : x.replace("0","na")) 
mapping.head(20)
```

```python
df = pd.merge(new_master, mapping ,how = "inner" , left_on="maincategory" , right_on="category_list")
df.head(10)
```

```python
df.drop(["category_list"], axis =1 , inplace=True)
```

```python
df.shape
```

Converting  wide dataframe to to Long using melt function

```python
value_vars = df.columns[10:]
id_vars = np.setdiff1d(df.columns,value_vars)

print(value_vars)
print(id_vars)
```

```python
long_df = pd.melt(df, id_vars=list(id_vars), value_vars= list(value_vars))
```

```python
long_df.shape
```

```python
id_vars.shape
```

```python
long_df.duplicated().sum()
```

```python
long_df = long_df[long_df["value"]==1]
long_df = long_df.drop("value", axis=1)
```

```python
long_df.shape
```

```python
#renaming the variable of column 
long_df = long_df.rename(columns={"variable":"sector"})
```

```python
long_df.info()
```

```python
df = long_df[(long_df["raised_amount_usd"] >= 5000000 ) & (long_df["raised_amount_usd"] <= 15000000 )]
```

```python
#group by sector company , and after calculate count and usm
df.groupby(["country_code" , "sector"]).raised_amount_usd.agg(["count","sum"])
```

```python
#ploting sector-wise count and sum of investment in the three countries
plt.figure(figsize=(25,20))

plt.subplot(2,1,1)
p = sns.barplot(x="sector", y= "raised_amount_usd", hue="country_code" , data=df)
p.set_xticklabels(p.get_xticklabels(),rotation=30)
plt.title("total investment amount(USD)")



plt.subplot(2,1,2)
p = sns.countplot(x="sector", hue="country_code" , data=df)
p.set_xticklabels(p.get_xticklabels(),rotation=30)
plt.title("number of investment ")
plt.legend(loc=1)


plt.show()
```
