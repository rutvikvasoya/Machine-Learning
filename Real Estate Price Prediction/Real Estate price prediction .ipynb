# Train - Test Spliting (Manually)

```python
import numpy as np
def split_train_test(data,ratio):
    np.random.seed(222)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * ratio)
    test_indicates = shuffled[:test_set_size]
    train_indicates = shuffled[test_set_size:]
    return data.iloc[train_indicates], data.iloc[test_indicates]

    
```

```python
train_set , test_set = split_train_test(housing,0.2)
print("size of train data set is",len(train_set))
print("size of test data set is",len(test_set))
```

# Creating Pipline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer


my_pipline = Pipeline([
    
    ("imputer",SimpleImputer(strategy="median")),
    ("std_scalet" , StandardScaler()),
    
])
```

```python
 housing_num_tr = my_pipline.fit_transform(housing_features)
```

```python
housing_num_tr
```

# Seclecting model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor




#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()



model.fit(housing_num_tr,housing_lables)
```

```python
some_data = housing_features.iloc[:5]
some_lables = housing_lables.iloc[:5]
```

```python
prepared_data = my_pipline.transform(some_data)
```

# Evaluting model

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
msc = mean_squared_error(housing_lables,housing_predictions)
rmsc = np.sqrt(msc)
```

```python
print(rmsc)
```

# Using better evalute technique - cross validation

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model,housing_num_tr,housing_lables, scoring="neg_mean_squared_error")
rmsc_score = np.sqrt(-score)
```
